```python
import random

# Komputer pilih nombor rawak antara 1 hingga 20
nombor_rahsia = random.randint(1, 20)
tekaan = 0

print("Jom main! Saya dah pilih satu nombor antara 1 hingga 20.")

# Gelung akan berjalan selagi tekaan salah
while tekaan != nombor_rahsia:
    tekaan = int(input("Masukkan tekaan anda: "))

    if tekaan < nombor_rahsia:
        print("Terlalu rendah! Cuba lagi.")
    elif tekaan > nombor_rahsia:
        print("Terlalu tinggi! Cuba lagi.")
    else:
        print("Tahniah! Tekaan anda betul! ")
```

```python
from IPython.display import HTML, display

# Kod gabungan Python + HTML5 + JavaScript untuk fungsi grafik bergerak
game_code = """
<div style="text-align: center; font-family: 'Arial', sans-serif; background: #f9f9f9; padding: 20px; border-radius: 15px; box-shadow: 0px 4px 10px rgba(0,0,0,0.1); width: 450px; margin: 0 auto;">

    <h2 style="color: #e65100; margin-bottom: 5px;">👦 Misi Upin & Ipin 🍗</h2>
    <p style="margin-top: 0; color: #555;">Dapatkan 5 ketul Ayam Goreng Kak Ros untuk menang!</p>

    <canvas id="padGame" width="400" height="400" style="border: 5px solid #ffb74d; background-color: #c8e6c9; border-radius: 10px; cursor: pointer;"></canvas>

    <div style="margin-top: 15px;">
        <h3 id="skorPaparan" style="color: #2e7d32; margin: 5px 0;">Ayam Diambil: 0 / 5</h3>
        <h2 id="mesejMenang" style="color: #d84315; display: none; font-weight: bold; animation: bounce 1s infinite;">🎉 BETUL, BETUL, BETUL! ANDA MENANG! 🎉</h2>
    </div>

    <p style="font-size: 12px; color: #777; margin-bottom: 5px;">💡 <i>Tip: Klik pada kotak hijau dahulu sebelum guna Keyboard (Arrow Keys / WASD)</i></p>

    <div style="margin-top: 10px;">
        <button onclick="gerak('atas')" style="padding: 10px 20px; font-size: 16px; background: #ff9800; color: white; border: none; border-radius: 5px; cursor: pointer; margin: 2px;">⬆️ Atas</button><br>
        <button onclick="gerak('kiri')" style="padding: 10px 20px; font-size: 16px; background: #ff9800; color: white; border: none; border-radius: 5px; cursor: pointer; margin: 2px;">⬅️ Kiri</button>
        <button onclick="gerak('kanan')" style="padding: 10px 20px; font-size: 16px; background: #ff9800; color: white; border: none; border-radius: 5px; cursor: pointer; margin: 2px;">➡️ Kanan</button><br>
        <button onclick="gerak('bawah')" style="padding: 10px 20px; font-size: 16px; background: #ff9800; color: white; border: none; border-radius: 5px; cursor: pointer; margin: 2px;">⬇️ Bawah</button>
    </div>
</div>

<script>
    const canvas = document.getElementById("padGame");
    const ctx = canvas.getContext("2d");

    // Tetapan Sistem Game
    let skor = 0;
    const sasaranSkor = 5;
    let gameTamat = false;

    // Posisi Watak Upin (Menggunakan Emoji sebagai Grafik)
    const pemain = {
        x: 180,
        y: 180,
        saiz: 35,
        kelajuan: 20,
        grafik: "👦"
    };

    // Posisi Ayam Goreng
    const ayam = {
        x: Math.floor(Math.random() * 340) + 20,
        y: Math.floor(Math.random() * 340) + 20,
        saiz: 35,
        grafik: "🍗"
    };

    // Fungsi untuk melukis grafik pada skrin
    function lukisGame() {
        // Bersihkan skrin lama
        ctx.clearRect(0, 0, canvas.width, canvas.height);

        // Lukis Rumput Hiasan (Sempadan dalam)
        ctx.strokeStyle = "#a5d6a7";
        ctx.lineWidth = 2;
        for(let i=0; i<canvas.width; i+=40) {
            ctx.beginPath(); ctx.moveTo(i, 0); ctx.lineTo(i, canvas.height); ctx.stroke();
            ctx.beginPath(); ctx.moveTo(0, i); ctx.lineTo(canvas.width, i); ctx.stroke();
        }

        // Lukis Ayam Goreng
        ctx.font = ayam.saiz + "px Arial";
        ctx.textBaseline = "top";
        ctx.fillText(ayam.grafik, ayam.x, ayam.y);

        // Lukis Pemain (Upin/Ipin)
        ctx.font = pemain.saiz + "px Arial";
        ctx.fillText(pemain.grafik, pemain.x, pemain.y);
    }

    // Fungsi mengesan jika Upin berjaya sentuh Ayam
    function semakPelanggaran() {
        let jarakX = Math.abs(pemain.x - ayam.x);
        let jarakY = Math.abs(pemain.y - ayam.y);

        // Jika jarak dekat, maksudnya ayam berjaya dimakan
        if (jarakX < 30 && jarakY < 30 && !gameTamat) {
            skor++;
            document.getElementById("skorPaparan").innerText = "Ayam Diambil: " + skor + " / " + sasaranSkor;

            if (skor >= sasaranSkor) {
                gameTamat = true;
                document.getElementById("mesejMenang").style.display = "block";
                pemain.grafik = "🤩"; // Tukar ekspresi muka Upin bila menang
            } else {
                // Tukar posisi ayam ke tempat rawak baru
                ayam.x = Math.floor(Math.random() * (canvas.width - 40)) + 10;
                ayam.y = Math.floor(Math.random() * (canvas.height - 40)) + 10;
            }
        }
    }

    // Logik Pergerakan
    function gerak(arah) {
        if (gameTamat) return;

        if (arah === "atas" && pemain.y > 0) pemain.y -= pemain.kelajuan;
        if (arah === "bawah" && pemain.y < canvas.height - pemain.saiz) pemain.y += pemain.kelajuan;
        if (arah === "kiri" && pemain.x > 0) pemain.x -= pemain.kelajuan;
        if (arah === "kanan" && pemain.x < canvas.width - pemain.saiz) pemain.x += pemain.kelajuan;

        semakPelanggaran();
        lukisGame();
    }

    // Mengesan input daripada Keyboard Laptop/PC
    window.addEventListener("keydown", function(e) {
        if(["ArrowUp","ArrowDown","ArrowLeft","ArrowRight","Space"].indexOf(e.code) > -1) {
            e.preventDefault(); // Mengelakkan skrin Colab daripada skrol naik/turun
        }

        if (e.code === "ArrowUp" || e.code === "KeyW") gerak("atas");
        if (e.code === "ArrowDown" || e.code === "KeyS") gerak("bawah");
        if (e.code === "ArrowLeft" || e.code === "KeyA") gerak("kiri");
        if (e.code === "ArrowRight" || e.code === "KeyD") gerak("kanan");
    });

    // Jalankan lukisan pertama kali apabila kod dimuatkan
    lukisGame();
</script>
"""

# Paparkan game di dalam output Google Colab
display(HTML(game_code))
```
